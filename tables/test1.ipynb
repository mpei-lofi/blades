```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import mglearn
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
tb = pd.read_excel(r'D:\SmirnovAOl\test1.xlsx')
target = np.array(tb[['profil','losses']])
data = np.array(tb[['t','alfaust','alfa1','Mah','alfa0', 'delta']])
target_names = np.array(['9009','9012','9015','9022','9038','5515','5520','4525','6030','6035','6520','7025'])
feature_names = np.array(['t','alfaust','alfa1','Mah','alfa0', 'delta'])
tb_dataset = {'data': data, 'target': target, 'target_names': target_names, 'feature_names': feature_names}
X_train,X_test,y_train,y_test=train_test_split(tb_dataset['data'], tb_dataset['target'], random_state=0)
reg = KNeighborsRegressor(n_neighbors=3)
reg.fit(X_train, y_train)
X_new = np.array([[0.7, 35, 13.4, 0.5, 90, 0.5]])
prediction = reg.predict(X_new)
print ('Прогноз: {}'.format(prediction))
y_pred = reg.predict(X_test)
#print ('Прогнозы для тествого набора: \n{}'.format(y_pred))
print("R^2 на тестовом наборе: {:.3f}".format(reg.score(X_test, y_test)))
print ('Правильность на тестовом наборе: {:.3f}'.format(np.mean(y_pred==y_test)))
print("Правильность на обучающем наборе: {:.3f}".format(reg.score(X_train, y_train)))
import seaborn as sns
print(sns.pairplot(tb, hue='profil', size=3))
```
